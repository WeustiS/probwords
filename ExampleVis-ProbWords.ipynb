```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
conv_arr = []
data_dict = {}
with open('text.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            line_count += 1
            for item in row:
                conv_arr.append(item)
                data_dict[item] = []
                
        else:
            i=0
            for item in row:
                data_dict[conv_arr[i]].append(item)
                i+=1
            line_count += 1
        if line_count == 25: # If you let it do the full dataset it gets pretty messy, so just for now we'll limit it :) I
            break
```

```python
plt.figure(figsize=(10,10)) 
for i in range(len(list(data_dict.values())[0])):
    for j in range(len(list(data_dict.values()))-3):
        #print(float(list(data_dict.values())[j][i]) + np.random.uniform(low=-.5, high=.5))
        # plt.plot(j+random.uniform(-.2, .2), float(list(data_dict.values())[j][i])+np.random.uniform(-.3,.3), 'or')
        degree = data_dict['Your Highest Level of Education'][i]
        if "PhD" in degree:
            plt.plot(range(0, 17), [float(x[i]) for x in list(data_dict.values())[:-3]], 'ro-', linewidth=.1)
        elif "MS" in degree:
            plt.plot(range(0, 17), [float(x[i]) for x in list(data_dict.values())[:-3]], 'bo-', linewidth=.1)
        else:
            plt.plot(range(0, 17), [float(x[i]) for x in list(data_dict.values())[:-3]], 'go-', linewidth=.1)
```
